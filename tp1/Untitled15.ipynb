```python
# Defino función dos gaussianas
def funcion(A, x):
    return A[0] * np.exp(- (x - A[1])**2 / (2 * A[2]**2))+ A[3] * np.exp(- (x - A[4])**2 / (2 * A[5]**2))
gaussiana = odr.Model(funcion)
```

```python


import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Definimos una función gaussiana con parámetros
# A[0] = amplitud, A[1] = centro, A[2] = ancho
def gauss(lam, A):
    amp, mu, sigma = A
    return amp * np.exp(-((lam - mu)/sigma)**2)

# Parámetros de los picos
A0_dttc = [0.6565, 762.9287, 20.3939]  # pico 0-0
A1_dttc = [0.464, 712.03, 65.68]  # pico 0-1


def calculoarea(A0, A1, xmin0, xmax0, xmin1, xmax1):
    # Integración
    area0, error0 = quad(gauss, xmin0, xmax0, args=(A0,))
    area1, error1 = quad(gauss, xmin1, xmax1, args=(A1,))

    # Graficar
    x_plot = np.linspace(min(xmin0, xmin1)-20, max(xmax0, xmax1)+20, 500)
    y0 = gauss(x_plot, A0)
    y1 = gauss(x_plot, A1)

    plt.plot(x_plot, y0, label="Pico 0-0")
    plt.plot(x_plot, y1, label="Pico 0-1")

    # Sombrear áreas integradas
    x_fill0 = np.linspace(xmin0, xmax0, 200)
    plt.fill_between(x_fill0, gauss(x_fill0, A0), alpha=0.3)

    x_fill1 = np.linspace(xmin1, xmax1, 200)
    plt.fill_between(x_fill1, gauss(x_fill1, A1), alpha=0.3, color="orange")

    plt.xlabel("λ")
    plt.ylabel("Absorbancia")
    plt.title("Gaussianas de dos picos")
    plt.legend()
    plt.grid(True)
    plt.show()

    print("Área pico 0-0:", area0)
    print("Área pico 0-1:", area1)
    print("Cociente I01/I00:", area1/area0)

    return area0, area1

# Llamamos a la función con límites de integración
calculoarea(A0_dttc, A1_dttc, xmin0=0, xmax0=800, xmin1=0, xmax1=900)




```
